```python
# checking for anagrams
# has time complexity - O(log n)
def anagram(s1 , s2):
    s1 = sorted(s1)
    s2 = sorted(s2)

    if s1 == s2:
        return True
    
    return False

print(anagram("listen","silent"))
```

```python
# another approch to check anagrams
# time complexity - O(n)

def second_anagram(s1,s2):
    if len(s1) != len(s2):
        return False
    
    count1 = {}
    count2 = {}
    
    for ch in s1:
        count1[ch] = count1.get(ch,0)+1
    for ch in s2:
        count2[ch] = count2.get(ch,0) + 1

    if count1 == count2:
        return True
    else :
        return False
    
print(second_anagram("listen","silent"))
```

```python
# remove duplicates and print new list of removed duplicates

def remove_deplicates(nums):
    
    s = set()
    result = []

    for ch in nums:
        if ch in s:
             continue
        else :
            s.add(ch)
            result.append(ch)

    return result

nums = [1,2,3,1,2]

print(remove_deplicates(nums))
```
