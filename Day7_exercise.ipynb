```python
# checking for anagrams
# has time complexity - O(n log n)
def anagram(s1 , s2):
    s1 = sorted(s1)
    s2 = sorted(s2)

    if s1 == s2:
        return True
    
    return False

print(anagram("listen","silent"))
```

```python
# another approch to check anagrams
# time complexity - O(n)

def second_anagram(s1,s2):
    if len(s1) != len(s2):
        return False
    
    count1 = {}
    count2 = {}
    
    for ch in s1:
        count1[ch] = count1.get(ch,0)+1
    for ch in s2:
        count2[ch] = count2.get(ch,0) + 1

    if count1 == count2:
        return True
    else :
        return False
    
print(second_anagram("listen","silent"))
```

```python
# remove duplicates and print new list of removed duplicates
# time complexity - O(n)
def remove_deplicates(nums):
    
    s = set()
    result = []

    for ch in nums:
        if ch in s:
             continue
        else :
            s.add(ch)
            result.append(ch)

    return result

nums = [1,2,3,1,2]

print(remove_deplicates(nums))
```

```python
# Time complexity - O(n^2) => explained step by step below in code 

def flatten_removeDplicate_ascending_list(nums):
    newList = []
    removedList = []

    # elements visiting exactly once - O (n)
    for ch in nums:
        for i in ch:
            newList.append(i)
            
    # return newList

    # here it becomes nested loop so - O(n^2)
    for i in newList:
        if i in removedList:
            continue
        else :
            removedList.append(i)

    # here for sorting time complexity is - O(n log(n))
    finalList = sorted(removedList) 
    # okay so here if i write sorted(removedList) only then it will not print sorted list
    # but if we create a variable and there we store sorted list then we can print that list by using tha variable name
    # sorted does not modify list , it creates new list , so we have to declare a variable and store it in it
    
    return finalList

    # SO OVERALL TIME COMPLEXITY IS = O(n) + O(n^2) + O(n log(n)) = take largest term = O(n^2)

nums = [[1,2],[9,4],[5,2],[1,8]]
print(flatten_removeDplicate_ascending_list(nums))
```

```python
def mostFrequent(nums):
    count = {}

    for ch in nums:
        count[ch] = count.get(ch,0)+1

    max_count = 0
    result = None

    for ch , count in count.items():
        
        if count > max_count:
            max_count = count
            result = ch

    return result

nums = [1,2,2,2,2,3,3,3,4]
print(mostFrequent(nums))
```

```python
# group words by length

def group_words_by_length(nums):
    result = {}

    for ch in nums :
        length = len(ch)

        if length not in result :
            result[length] = []

        result[length].append(ch)

    return result

nums = ["cat", "dog", "elephant", "bat"]
print(group_words_by_length(nums))
```

```python
# find missing number

def missing(nums):
    n = len(nums) + 1

    expected_sum = n*(n+1)//2
    current_sum = sum(nums)

    return expected_sum - current_sum


nums = [1,2,3,5]
print(missing(nums))
```

```python
# another approch using loops to find missing element
def missing2(nums):
    n = len(nums)+1

    for i in range(1,n+1):
        if i not in nums:
            return i
        
nums = [1,2,3,5]
print(missing2(nums))
```

```python
# merge two sorted list 

def merge_sorted_list(list1,list2):
    result = []
    i = j = 0

    while i  < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i +=1

        else :
            result.append(list2[j])
            j+=1

    while i < len(list1):
        result.append(list1[i])
        i = i+1
    while( j < len(list2)):
        result.append(list2[j])
        j = j+1

    return result

list1 = [1,3,6]
list2 = [2,4,5]

print(merge_sorted_list(list1,list2))
```

```python
# Two sum

# brute force approach - O(n^2)

def find_pair(nums,target):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[i] + nums[j] == target:
                return i ,j

nums = [2, 7, 11, 15]
target = 9

print(find_pair(nums,target))
```

```python
# optimized version

def find_pair1(nums,target):
    seen = {}

    for i , num in enumerate(nums):
        complement = target - num

        if complement in seen:
            return [seen[complement],i]
        
        seen[num] = i


nums = [2, 7, 11, 15]
target = 9

print(find_pair1(nums,target))
```
