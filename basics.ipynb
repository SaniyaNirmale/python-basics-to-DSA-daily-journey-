```python
def is_palindrome(s):
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left+=1
        right-=1

    return True


is_palindrome("madam")
        
```

```python
def duplicates(nums):
    freq = {}

    for i in nums:
        freq[i] = freq.get(i,0)+1

    for i,count in freq.items():
        if count>1:
            print(i)

nums = [1,2,4,1,7,2,1,4]
duplicates(nums)

        
```

```python
def flatten(nums):
    result = []

    for i in nums:
        for item in i:
            result.append(item)
    return result

nums = [[1,2],[3,4]]
flatten(nums)
```

```python
def second_largest(nums):
    if nums[0]>nums[1]:
        largest , second = nums[0],nums[1]
    else:
        largest , second = nums[1],nums[0]


    for i in nums:
        if i > largest:
            second = largest
            largest = i
        elif i > second & i!=largest:
            second = i
    return second

nums = [1,6,3,9,4]

second_largest(nums)
```

```python
def anagram (s1,s2):
    s1 = sorted(s1)
    s2 = sorted(s2)

    if s1 == s2:
        return True
    
    return False

anagram("silent","listen")
```

```python
def anagram1(s1,s2):
    if len(s1) != len(s2):
        return False
    
    count1 = {}
    count2 = {}

    for ch in s1:
        count1[ch] = count1.get(ch,0)+1
    for ch in s2:
        count2[ch] = count2.get(ch,0)+1

    if count1 == count2:
        return True
    return False


anagram1("silent","listen")
```

```python
def remove_duplicates(nums):
    s = set()
    result = []

    for ch in nums:
        if ch in s:
            continue
        else :
            s.add(ch)
            result.append(ch)

    return result

nums = [1,5,2,5,3,2,1]
remove_duplicates(nums)
```
